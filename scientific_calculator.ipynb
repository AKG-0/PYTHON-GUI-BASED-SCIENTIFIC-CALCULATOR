```python
#import tkinter module
from tkinter import *
import math
import tkinter
window=tkinter.Tk()
window.title("SCIENTIFIC CALCULATOR")

value = ""
def click(num):
    global value
    value=value + str(num)
    insideText.set(value)
def equal():
    try:
        global value
        value=str(eval(value))
        insideText.set(value)
    except:
        insideText.set("ERROR")
        value=""
def clearall():
    global value
    value=""
    insideText.set(value)
def clear1():
    global value
    value=value[:-1]
    insideText.set(value)
def e():
    global value
    value=value+str(math.e)
    insideText.set(value)
def pi():
    global value
    value=value+str(math.pi)
    insideText.set(value)
def log():
    global value
    value=math.log(float(txtDisplay.get()))
    insideText.set(value)
def log10():
    global value
    value=math.log10(float(txtDisplay.get()))
    insideText.set(value)
def cos():
    global value
    value=math.cos(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def sin():
    global value
    value=math.sin(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def tan():
    global value
    value=math.tan(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def cosh():
    global value
    value=math.cosh(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def sinh():
    global value
    value=math.sinh(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def tanh():
    global value
    value=math.tanh(math.radians(float(txtDisplay.get())))
    insideText.set(value)
def mathsPM():
    global value
    value=str(-(float(txtDisplay.get())))
    insideText.set(value)
def degrees():
    global value
    value=math.degrees(float(txtDisplay.get()))
    insideText.set(value)
def exp():
    global value
    value=math.exp(float(txtDisplay.get()))
    insideText.set(value)
#widges are added here
insideText=StringVar()

txtDisplay=Entry(window,textvariable=insideText,font=('arial',10,'bold'),bd=20,width=30,justify=RIGHT)
txtDisplay.grid(row=1,column=0,columnspan=6)
txtDisplay.insert(0,"0")
Button(window,text='sinh',width='5',command=sinh).grid(row=2,column=0)
Button(window,text='sin',width='5',command=sin).grid(row=2,column=1)
Button(window,text='π',width='5',command=pi).grid(row=2,column=2)
Button(window,text='√',width='5',command=lambda:click('**0.5')).grid(row=2,column=3)
Button(window,text='CA',width='5',command=clearall).grid(row=2,column=4)
Button(window,text='C',width='5',command=clear1).grid(row=2,column=5)
Button(window,text=chr(247),width='5',command=lambda:click('/')).grid(row=2,column=6)

Button(window,text='cosh',width='5',command=cosh).grid(row=3,column=0)
Button(window,text='cos',width='5',command=cos).grid(row=3,column=1)
Button(window,text='e',width='5',command=e).grid(row=3,column=2)
Button(window,text='7',width='5',command=lambda:click(7)).grid(row=3,column=3)
Button(window,text='8',width='5',command=lambda:click(8)).grid(row=3,column=4)
Button(window,text='9',width='5',command=lambda:click(9)).grid(row=3,column=5)
Button(window,text='*',width='5',command=lambda:click('*')).grid(row=3,column=6)


Button(window,text='tanh',width='5',command=tanh).grid(row=4,column=0)
Button(window,text='tan',width='5',command=tan).grid(row=4,column=1)
Button(window,text='±',width='5',command=mathsPM).grid(row=4,column=2)
Button(window,text='4',width='5',command=lambda:click(4)).grid(row=4,column=3)
Button(window,text='5',width='5',command=lambda:click(5)).grid(row=4,column=4)
Button(window,text='6',width='5',command=lambda:click(6)).grid(row=4,column=5)
Button(window,text='-',width='5',command=lambda:click('-')).grid(row=4,column=6)

Button(window,text='log10',width='5',command=log10).grid(row=5,column=0)
Button(window,text='ln',width='5',command=log).grid(row=5,column=1)
Button(window,text='deg',width='5',command=degrees).grid(row=5,column=2)
Button(window,text='1',width='5',command=lambda:click(1)).grid(row=5,column=3)
Button(window,text='2',width='5',command=lambda:click(2)).grid(row=5,column=4)
Button(window,text='3',width='5',command=lambda:click(3)).grid(row=5,column=5)
Button(window,text='+',width='5',command=lambda:click('+')).grid(row=5,column=6)

Button(window,text='Exp',width='5',command=exp).grid(row=6,column=0)
Button(window,text='(',width='5',command=lambda:click('(')).grid(row=6,column=1)
Button(window,text=')',width='5',command=lambda:click(')')).grid(row=6,column=2)
Button(window,text='^',width='5',command=lambda:click('**')).grid(row=6,column=3)
Button(window,text='0',width='5',command=lambda:click(0)).grid(row=6,column=4)
Button(window,text='.',width='5',command=lambda:click('.')).grid(row=6,column=5)
Button(window,text='=',width='5',command=equal).grid(row=6,column=6)

window.mainloop()
```
